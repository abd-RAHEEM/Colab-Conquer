Load and Preprocess

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data
url = "https://bit.ly/4nejNue"
df = pd.read_csv(url)

# Preview
print(df.head())

# Check data types
print(df.dtypes)

# Basic checks
print("Shape:", df.shape)
print("Missing values:\n", df.isnull().sum())
```

 Pandas Statistical Functions including basic funtions

```python
# Summary statistics
print(df.describe())

# All statistics one by one
print("Mean:\n", df.mean(numeric_only=True))
print("Median:\n", df.median(numeric_only=True))
print("Mode:\n", df.mode(numeric_only=True))
print("Min:\n", df.min(numeric_only=True))
print("Max:\n", df.max(numeric_only=True))
print("Standard Deviation:\n", df.std(numeric_only=True))
print("Variance:\n", df.var(numeric_only=True))
print("Skewness:\n", df.skew(numeric_only=True))
print("Kurtosis:\n", df.kurt(numeric_only=True))
print("Sum:\n", df.sum(numeric_only=True))
print("Count:\n", df.count())
print("Correlation:\n", df.corr(numeric_only=True))
print("Covariance:\n", df.cov(numeric_only=True))
```

Matplotlib Visualizations

1. Simple Line Plot

```python
plt.figure()
plt.plot(df['sepal_length'], label='Sepal Length')
plt.plot(df['sepal_width'], label='Sepal Width')
plt.title("Simple Line Plot of Sepal Dimensions")
plt.xlabel("Sample Index")
plt.ylabel("Measurement (cm)")
plt.legend()
plt.grid(True)
plt.show()
```

2. Scatter Plot

```python
plt.figure()
plt.scatter(df['sepal_length'], df['petal_length'], c='r')
plt.title("Scatter: Sepal vs Petal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Petal Length")
plt.grid(True)
plt.show()
```

3. Visualizing Errors

```python
mean = df['sepal_length'].mean()
std = df['sepal_length'].std()
x = df.index
y = df['sepal_length']

plt.figure()
plt.errorbar(x, y, yerr=std, fmt='o', ecolor='red', capsize=2)
plt.title("Error Bar: Sepal Length with STD")
plt.xlabel("Index")
plt.ylabel("Sepal Length")
plt.grid(True)
plt.show()
```

4. Density Plot (Approximated using Hist with density=True)

```python
plt.figure()
plt.hist(df['petal_length'], bins=20, density=True, color='skyblue')
plt.title("Density Plot of Petal Length")
plt.xlabel("Petal Length")
plt.ylabel("Density")
plt.grid(True)
plt.show()
```

5. 2D Contour Plot (Using plt.contour)

```python
import numpy as np

x = df['sepal_length']
y = df['sepal_width']

X, Y = np.meshgrid(np.linspace(x.min(), x.max(), 50), np.linspace(y.min(), y.max(), 50))
Z = np.sin(X) + np.cos(Y)  # synthetic function just to demonstrate contour

plt.figure()
cp = plt.contour(X, Y, Z, cmap='viridis')
plt.title('Contour Plot (Synthetic)')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')
plt.colorbar(cp)
plt.show()
```

6. Histogram with Binning

```python
plt.figure()
plt.hist(df['petal_width'], bins=[0, 0.5, 1, 1.5, 2, 2.5], edgecolor='black')
plt.title("Histogram with Binning - Petal Width")
plt.xlabel("Petal Width")
plt.ylabel("Frequency")
plt.grid(True)
plt.show()
```

7. Multiple Subplots

```python
fig, axs = plt.subplots(2, 2, figsize=(10, 8))

axs[0, 0].hist(df['sepal_length'], color='orange')
axs[0, 0].set_title('Sepal Length')

axs[0, 1].hist(df['sepal_width'], color='green')
axs[0, 1].set_title('Sepal Width')

axs[1, 0].hist(df['petal_length'], color='blue')
axs[1, 0].set_title('Petal Length')

axs[1, 1].hist(df['petal_width'], color='red')
axs[1, 1].set_title('Petal Width')

plt.tight_layout()
plt.show()
```

8. 3D Plotting with Matplotlib

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df['sepal_length'], df['sepal_width'], df['petal_length'], c='purple')
ax.set_xlabel('Sepal Length')
ax.set_ylabel('Sepal Width')
ax.set_zlabel('Petal Length')
plt.title("3D Plot of Iris Features")
plt.show()
```
